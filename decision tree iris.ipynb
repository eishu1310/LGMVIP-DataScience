# Importing Libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
from sklearn import metrics
```

```python
df=pd.read_csv('iris.csv')
```

# Determining target & feature variables

```python
feature_cols=['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
       
x=df[feature_cols]
y=df.species
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

# Decision Tree Classifier

```python
clf = DecisionTreeClassifier() #creating decision tree classifier object 
```

```python
clf = clf.fit(x_train,y_train) #training 
```

```python
y_pred = clf.predict(x_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))  #printing accuracy
```

# Decision Tree Graphical Representation

```python
import matplotlib.pyplot as plt
```

```python
model_all_params= DecisionTreeClassifier().fit(x_train,y_train)
plt.figure(figsize=(11,6))
tree.plot_tree(model_all_params,filled=True)
plt.show()
```
